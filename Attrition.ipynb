## 6.3 Checking and Removal of Outliers

### 6.3.1 Checking for Outliers

```python
plt.figure(figsize=(50,45))
iterator=1

for i in df.columns:
    if iterator <=34:
        ax=plt.subplot(7, 5, iterator)
        sns.boxplot(df[i])
        plt.xlabel(i, fontsize=10)
        plt.ylabel('Frequency', fontsize=10)
    iterator +=1
plt.tight_layout()
```

### 6.3.2 Removal of Outliers using Interquantile method

```python
for i in ['MonthlyIncome','NumCompaniesWorked','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']:
    q75,q25=np.percentile(df.loc[:, i],[75,25])
    intr_qr=q75-q25
    max=q75 + (1.5*intr_qr)
    min=q25 - (1.5*intr_qr)
    df.loc[df[i]<min,i]=np.nan
    df.loc[df[i]>max,i]=np.nan
```

```python
df.isnull().sum()
```

```python
# Using Simple Imputer to missing data
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy="most_frequent")
list = ['MonthlyIncome','NumCompaniesWorked','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']
for i in list:
    df[i]=imp.fit_transform(df[i].values.reshape(-1,1))
```

```python
# To check all the nullvalues are filled 
df.isnull().sum()
```

## 6.4 Taking x & y from table and transforming using PCA Class

```python
x=df.drop(df.columns[1], axis=1)
x
```

```python
x.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=25)
```

```python
y=df['Attrition']
```

```python
xnew=pca.fit_transform(x)
```

```python
xnew.shape
```

```python
pd.DataFrame(data=xnew)
```

## 6.5 Creating train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(xnew, y, test_size=0.2, random_state=45)
```

## 6.6 Passing All Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
model = [LogisticRegression(), GaussianNB(), SVC(), DecisionTreeClassifier(), RandomForestClassifier(), ExtraTreesClassifier(), GradientBoostingClassifier(), AdaBoostClassifier(), KNeighborsClassifier()]
for m in model:
    m.fit(x_train,y_train)
    predm=m.predict(x_test)
    print("Accuracy Score of",m,"is:", accuracy_score(y_test,predm))
    print("Confusion Matrix of", m, "is:", confusion_matrix(y_test,predm))
    print("Classification Report of", m, "is:", classification_report(y_test,predm))
    print('\n')
```

From above we can see that GaussianNB has highest accuracy score

 ## 7. cross_val_score

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
import numpy as np

gnb = GaussianNB()

cv_scores = cross_val_score(gnb, xnew, y, cv=5)

# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean CV score:", np.mean(cv_scores))
print("Standard deviation of CV scores:", np.std(cv_scores))
```

## 8. Selecting Final Model

```python
x_train,x_test,y_train,y_test=train_test_split(xnew,y,test_size=0.2,random_state=45)
FinalModel=GaussianNB()
FinalModel.fit(x_train,y_train)
pred=FinalModel.predict(x_test)
acc=accuracy_score(y_test,pred)
print(f"Accuracy Score: {acc*100}")
```
