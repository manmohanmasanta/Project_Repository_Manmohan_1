## 5.3 Checking and Removal of Outliers

### 5.3.1 Checking for outliers

```python
plt.figure(figsize=(20,15))
iterator=1
for i in df.columns:
    if iterator <=17:
        ax=plt.subplot(4,5, iterator)
        sns.boxplot(df[i])
        plt.xlabel(i, fontsize=10)
        plt.ylabel('Frequency', fontsize=10)
    iterator +=1
plt.tight_layout()
```

We can drop E, it is having outliers and having low correlation with Target variable as well. <br>
Rest all features have negligible or no outliers.

### 5.3.2 Removal of outliers

```python
# Removal of outliers from R, ERA, SHO, SV

for i in ['R','ERA', 'SHO', 'SV']:
    q75,q25=np.percentile(df.loc[:,i],[75,25])
    intr_qr=q75-q25
    
    max=q75+(1.5*intr_qr)
    min=q25-(1.5*intr_qr)
    
    df.loc[df[i]<min,i]=np.nan
    df.loc[df[i]>max,i]=np.nan
```

```python
df.isnull().sum()
```

There are null values present in the dataset.<br> We will use Mean, Median, Mode to fill the null value.

```python
#Filling the null values with Median of R

Median_R=df['R'].median()
print("Median of R:", Median_R)
df['R']=df['R'].fillna(Median_R)
```

```python
#Filling the null values with Median of R

Mean_ERA=df['ERA'].mean()
print("Mean of ERA:", Mean_ERA)
df['ERA']=df['ERA'].fillna(Mean_ERA)
```

```python
#Filling the null values with Median of SHO

Median_SHO=df['SHO'].median()
print("Median of SHO:", Median_SHO)
df['SHO']=df['SHO'].fillna(Median_SHO)
```

```python
#Filling the null values with Median of SV

Median_SV=df['SV'].median()
print("Median of SV:", Median_SV)
df['SV']=df['SV'].fillna(Median_SV)
```

```python
df.isnull().sum()
```

```python
#Checking for outliers again after removal.

plt.figure(figsize=(20,15))
iterator=1
for i in df.columns:
    if iterator <=17:
        ax=plt.subplot(4,5, iterator)
        sns.boxplot(df[i])
        plt.xlabel(i, fontsize=10)
        plt.ylabel('Frequency', fontsize=10)
    iterator +=1
plt.tight_layout()
```

We can remove columns : AB, H, HR, BB, SO, CG, E ;since it has too many outliers in it.<br>
We need to keep columns: W, R, 2B, 3B, SB, RA, ER, ERA, SV

```python
df.drop(['AB','H','HR','BB','SO','CG','3B','E','SHO'],axis=1, inplace=True)
df
```

## 5.4 Checking and Removal of Skewness

```python
df.skew()
```

Only very small skewness is present in Column SV

We will use different transformation technique to remove skewness from Column SV

```python
#Finding the best possible transformation

funclist=[np.cbrt,np.log,np.sqrt,np.exp]
column_list=['SV']

skewness_output=[]

for column in column_list:
    row=[column]
    for func in funclist:
        x=func(df[column])
        skewness=x.skew()
        row.append(skewness)
    skewness_output.append(row)
    
skewness_df=pd.DataFrame(skewness_output, columns=['Column','Cbrt','Log','Sqrt','Exp'])

print(skewness_df)
```

```python
#Performing transformation

df['SV']=np.log(df['SV'])
```

# 6. Model Preparation 

## 6.1 Selecting Features and Labels

```python
x=df.drop('W', axis=1)
y=df['W']
```

## 6.2 Finding the best Random State Model

```python
#Importing Train_Test_Model and LinearRegression
#Importing metrics
from sklearn.linear_model import LinearRegression
maxScore=0

# 1.Splitting the features(x), in training and testing set with 80% data being sent to training.
# 2. Applying Standard Scaler on training set only.

for i in range(201):
    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = i)
    from sklearn.preprocessing import StandardScaler
    SC = StandardScaler()
    x_train = SC.fit_transform(x_train)
    LR = LinearRegression()
    LR.fit(x_train,y_train)
    mscore=LR.score(SC.transform(x_test),y_test)
    if mscore>maxScore:
        maxScore=mscore
        randomState=i
    
print('Best Score = {} For Random state = {}'.format(maxScore*100,randomState))
```

## 6.3 Findling the Best Peforming model using the random state

```python
r2_Score=[]
Mean_Absolute_Error=[]
Mean_Squared_Error=[]

#1. LinearRegression (with best random_state)

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=165)
SC = StandardScaler()
x_train = SC.fit_transform(x_train)
LR = LinearRegression()
LR.fit(x_train,y_train)
predLR = LR.predict(SC.transform(x_test))
r2 = r2_score(y_test, predLR)
mae = mean_absolute_error(y_test, predLR)
mse = mean_squared_error(y_test, predLR)

print('R2 SCore:', r2)
print('Mean Absolute Error:', mae)
print('Mean Sqaured Error:', mse)
```

# 7.Hyper Parameter Tuning

## 7.1 For Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

param_grid = {
    'fit_intercept': [True, False],
    'n_jobs': [None, 1, 2, 3],  
    'positive': [False],
}

LR = LinearRegression()

grid_search = GridSearchCV(LR, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x, y)

best_params = grid_search.best_params_
print(f"Best Parameters : {best_params}")

best_model = grid_search.best_estimator_
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

params = {
    'fit_intercept': False,
    'n_jobs': None,
    'positive': False,
}

LR = LinearRegression(fit_intercept=False, n_jobs=None, positive=False)

LR.fit(x, y)

y_pred = LR.predict(x)

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Used Parameters : {params}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mse)}")
print(f"R-squared (R2) Score: {r2}")
```

```python
mae_values=[]
mse_values=[]
rmse_values=[]
r2score_values=[]

mae_values.append(mae)
mse_values.append(mse)
rmse_values.append(np.sqrt(mse))
r2score_values.append(r2)

print(mae_values)
print(mse_values)
print(rmse_values)
print(r2score_values)
```
