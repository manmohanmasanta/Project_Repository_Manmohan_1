## 5.3 Checking and Removal of Outliers

### 5.3.1 Checking for outliers

```python
plt.figure(figsize=(20,15))
iterator=1
for i in df.columns:
    if iterator <=17:
        ax=plt.subplot(4,5, iterator)
        sns.boxplot(df[i])
        plt.xlabel(i, fontsize=10)
        plt.ylabel('Frequency', fontsize=10)
    iterator +=1
plt.tight_layout()
```

We can drop E, it is having outliers and having low correlation with Target variable as well. <br>
Rest all features have negligible or no outliers.

### 5.3.2 Removal of outliers

```python
# Removal of outliers from R, ERA, SHO, SV

for i in ['R','ERA', 'SHO', 'SV']:
    q75,q25=np.percentile(df.loc[:,i],[75,25])
    intr_qr=q75-q25
    
    max=q75+(1.5*intr_qr)
    min=q25-(1.5*intr_qr)
    
    df.loc[df[i]<min,i]=np.nan
    df.loc[df[i]>max,i]=np.nan
```

```python
df.isnull().sum()
```

There are null values present in the dataset.<br> We will use Mean, Median, Mode to fill the null value.

```python
#Filling the null values with Median of R

Median_R=df['R'].median()
print("Median of R:", Median_R)
df['R']=df['R'].fillna(Median_R)
```

```python
#Filling the null values with Median of R

Mean_ERA=df['ERA'].mean()
print("Mean of ERA:", Mean_ERA)
df['ERA']=df['ERA'].fillna(Mean_ERA)
```

```python
#Filling the null values with Median of SHO

Median_SHO=df['SHO'].median()
print("Median of SHO:", Median_SHO)
df['SHO']=df['SHO'].fillna(Median_SHO)
```

```python
#Filling the null values with Median of SV

Median_SV=df['SV'].median()
print("Median of SV:", Median_SV)
df['SV']=df['SV'].fillna(Median_SV)
```

```python
df.isnull().sum()
```

## 5.4 Checking and Removal of Skewness

```python
df.skew()
```

Only very small skewness is present in Column SV

We will use different transformation technique to remove skewness from Column SV and Column E

```python
#Finding the best possible transformation

funclist=[np.cbrt,np.log,np.sqrt,np.exp]
column_list=['SV']

skewness_output=[]

for column in column_list:
    row=[column]
    for func in funclist:
        x=func(df[column])
        skewness=x.skew()
        row.append(skewness)
    skewness_output.append(row)
    
skewness_df=pd.DataFrame(skewness_output, columns=['Column','Cbrt','Log','Sqrt','Exp'])

print(skewness_df)
```

```python
#Performing transformation

df['SV']=np.log(df['SV'])
```

## 6.2 Finding the best Random State Model

```python
#Importing Train_Test_Model and LinearRegression
#Importing metrics
from sklearn.linear_model import LinearRegression
maxScore=0

# 1.Splitting the features(x), in training and testing set with 80% data being sent to training.
# 2. Applying Standard Scaler on training set only.

for i in range(201):
    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = i)
    from sklearn.preprocessing import StandardScaler
    SC = StandardScaler()
    x_train = SC.fit_transform(x_train)
    LR = LinearRegression()
    LR.fit(x_train,y_train)
    mscore=LR.score(SC.transform(x_test),y_test)
    if mscore>maxScore:
        maxScore=mscore
        randomState=i
    
print('Best Score = {} For Random state = {}'.format(maxScore*100,randomState))
```

## 6.4 Using K Fold Cross Validation score to find the alpha value

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_scores=[]
rmse_score=[]
Mean_Absolute_Error=[]
std_dev_rmse=[]
std_dev_r2=[]

#1. Cross_val_score of LinearRegression 

kf=KFold(n_splits=5, shuffle=True, random_state=165)
LinearCV=LinearRegression()
SC=StandardScaler()
x_scaled=SC.fit_transform(x)
r2_score=cross_val_score(LinearCV, x_scaled, y, cv=kf, scoring='r2')
cv_scores=cross_val_score(LinearCV, x_scaled, y, cv=kf, scoring='neg_mean_absolute_error')
mean_absolute_scores=np.mean(np.absolute(cv_scores))
rmse_scores=np.sqrt(-cv_scores)

print(f"Mean Absolute Error: {mean_absolute_scores}")
print(f"Mean RMSE: {np.mean(rmse_scores)}")
print(f"Standard Deviation of RMSE: {np.std(rmse_scores)}")
print(f"Mean R2 Score:{np.mean(r2_score)*100}")
print(f"Standard Deviation of R2 Score:{np.std(r2_score)}")
print()
```

```python
r2_scores.append(np.mean(r2_score)*100)
Mean_Absolute_Error.append(mean_absolute_scores)
rmse_score.append(np.mean(rmse_scores))
std_dev_rmse.append(np.std(rmse_scores))
std_dev_r2.append(np.std(r2_score))
print(r2_scores)
print(Mean_Absolute_Error)
print(rmse_score)
print(std_dev_rmse)
print(std_dev_r2)
```
