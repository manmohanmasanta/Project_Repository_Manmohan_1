## 6.5 Removing Outliers

```python
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df.iloc[:, 0:10]))
z
```

```python
threshold=3
```

```python
new_df = df[z<threshold]
```

```python
new_df
```

new dataframe with outliers removed as shown.

```python
new_df.isnull().sum()
```

After removing outliers, null values are present in multiple columns. We need to fill the null value with mean

```python
#To find mean of Standard Error
StandardError_mean = new_df['Standard Error'].mean()
print(StandardError_mean)
```

```python
#To find mean of Family
Family_mean = new_df['Family'].mean()
print(Family_mean)
```

```python
#To find mean of Trust (Government Corruption)
TrustGovernmentCorruption_mean = new_df['Trust (Government Corruption)'].mean()
print(TrustGovernmentCorruption_mean)
```

```python
#To find mean of Generosity
Generosity_mean = new_df['Generosity'].mean()
print(Generosity_mean)
```

```python
#To find mean of Dystopia Residual
DystopiaResidual_mean = new_df['Dystopia Residual'].mean()
print(DystopiaResidual_mean)
```

```python
#Filling NaN with Mean
new_df['Standard Error'] = new_df['Standard Error'].fillna(StandardError_mean)
```

```python
new_df['Family'] = new_df['Family'].fillna(Family_mean)
```

```python
new_df['Trust (Government Corruption)'] = new_df['Trust (Government Corruption)'].fillna(TrustGovernmentCorruption_mean)
```

```python
new_df['Generosity'] = new_df['Generosity'].fillna(Generosity_mean)
```

```python
new_df['Dystopia Residual'] = new_df['Dystopia Residual'].fillna(DystopiaResidual_mean)
```
