# 3. Finding null values

```python
#Before converting the categorical columns to numerical, we need to find the null values and assign any value to NaN
df.isnull().sum()
```

```python
for i in df.columns:
    if df[i].isnull().any():
        df[i].fillna(-1,inplace=True)
df
```

1. We find if NaN values are present in range of columns<br>
2. We fill the NaN values with '-1'.

```python
# To check for null values after filling NaN
df.isnull().sum()
```

No null values are present in any column, now we can go ahead and convert the columns using encoding technique

# 3. Encoding the Categorical columns

```python
df.dtypes
```

CGPA column is float type<br>
All other columns are object types

```python
from sklearn.preprocessing import OrdinalEncoder
OE=OrdinalEncoder()
for i in df.columns[1:42]:
    df[i]=df[i].astype(str)

for i in df.columns[1:42]:
    df[i]=OE.fit_transform(df[i].values.reshape(-1,1))
df
```

Column 'Seat No' has all the unique values and hence can be dropped
